```python
from autoscraper import AutoScraper
url="https://www.amazon.in/s?k=redmi+phones"
wanted_list = ["₹6,799","Redmi 9A (Nature Green, 2GB Ram, 32GB Storage) | 2GHz Octa-core Helio G25 Processor"," 36,479"]
```

```python
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_34pe': 'Title', 'rule_4s44': 'Rating'})
scraper.keep_rules(['rule_34pe','rule_4s44'])
scraper.save('amazon-search')
```

```python
results = scraper.get_result_similar("https://www.amazon.in/s?k=iphone",group_by_alias=True)
```
